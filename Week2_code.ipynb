```python
# 2.8.2
def eulerian_cycle(g: Dict[int, List[int]]) -> Iterable[int]:
    """Constructs an Eulerian cycle in a graph."""
    # cycle stores the final version of the Eulerian cycle
    cycle = []
    # retrieve the first key as the initial start point
    startPoint = next(iter(g))
    while len(g.keys()) > 0 :
        # a tentative cycle that probably doesn't walk the whole graph
        cycle0 = [startPoint]
        # start from the start point, add the next node to cycle0
        # and delete each element that has been walked in the cycle
        cycle0.append(g[startPoint][0])
        
        if len(g[startPoint]) == 1:
                del g[startPoint]
        else:
                g[startPoint].remove(g[startPoint][0])
        # cycle0 will keep updating until it reach the start point again
        # the update will always add the node that the last element of the cycle list point to
        while cycle0[-1] in g.keys():
            parent = cycle0[-1]
            cycle0.append(g[parent][0])
            if len(g[parent]) == 1:
                del g[parent]
            else:
                g[parent].remove(g[parent][0])
  
               
        # incorporate with formal cycles
        cycle.extend(cycle0)
        # find a new start point that has unused edges
        index = 0
        for i, key in enumerate(cycle):
            if key in g.keys():
                startPoint = key
                index = i
        
        # walk the tentative cycle again from the new startPoint, update the order of node
        newCycle = []
        newCycle.extend(cycle[index:])
        if index != 1:
            newCycle.extend(cycle[1:index])
        
        cycle = newCycle
        
    return cycle
```
